```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
df = pd.read_csv("../crime-2020.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
df['Date'] = df['DATE OCC'].dt.strftime('%d')
```

```python
df['Month'] = df['DATE OCC'].dt.strftime('%m')
df['Year'] = df['DATE OCC'].dt.strftime('%Y')
```

```python
df['Time'] = df['DATE OCC'].dt.strftime('%I:%M:%S %p')
```

```python
df.head()
df.tail()
```

```python
df_encoded = pd.get_dummies(df, columns=['Vict Sex', 'Vict Descent'], dtype=int)

numerical_df = df_encoded.select_dtypes(include=['number'])
corr_matrix = numerical_df.corr()
plt.figure(figsize=(24, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Numerical Correlation')
plt.show()
plt.show()
```

```python
area_crm_counts = df.groupby('AREA')['Crm Cd'].count().reset_index()
colormap = cm.viridis(np.linspace(0, 1, len(area_crm_counts['AREA'])))

# Sort the areas for better visualization (optional)
area_crm_counts = area_crm_counts.sort_values('Crm Cd', ascending=False)

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(area_crm_counts['AREA'].astype(str), area_crm_counts['Crm Cd'], color=colormap)
plt.xlabel('Area (Numerical Code)')
plt.ylabel('Count of Crime Codes')
plt.title('Count of Crime Codes by Area')
plt.xticks(rotation=45)  # Rotate the x labels for better readability
plt.show()
```

```python
month_crm_counts = df.groupby('Month')['Crm Cd'].count().reset_index()
colormap = cm.viridis(np.linspace(0, 1, len(month_crm_counts['Month'])))

# Sort the areas for better visualization (optional)
month_crm_counts = month_crm_counts.sort_values('Crm Cd', ascending=False)

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(month_crm_counts['Month'].astype(str), month_crm_counts['Crm Cd'], color=colormap)
plt.xlabel('Month')
plt.ylabel('Count of Crime Codes')
plt.title('Count of Crime Codes by Month')
plt.xticks(rotation=45)  # Rotate the x labels for better readability
plt.show()
```

```python
gender_counts_by_area = df_encoded.groupby('AREA')[['Vict Sex_M', 'Vict Sex_F']].sum().reset_index()

positions = np.arange(len(gender_counts_by_area['AREA']))
width = 0.4  # the width of the bars

male_color = 'skyblue'
female_color = 'pink'
plt.figure(figsize=(14, 8))

# Plot bars for males
plt.bar(positions - width/2, gender_counts_by_area['Vict Sex_M'], width=width, label='Male', color=male_color)

# Plot bars for females
plt.bar(positions + width/2, gender_counts_by_area['Vict Sex_F'], width=width, label='Female', color=female_color)
plt.xlabel('AREA')
plt.ylabel('Number of Victims')
plt.title('Number of Victims by Gender and Area')
plt.xticks(positions, gender_counts_by_area['AREA'].astype(str), rotation=45)
plt.legend()

plt.show()
```

```python
descent_counts_by_area = df_encoded.groupby('AREA')[['Vict Descent_W', 'Vict Descent_B', 'Vict Descent_H']].sum().reset_index()

positions = np.arange(len(descent_counts_by_area['AREA']))
width = 0.4  # the width of the bars

plt.figure(figsize=(14, 8))

# Plot bars for white people
plt.bar(positions - width/3, descent_counts_by_area['Vict Descent_W'], width=width, label='White', color='#A9CCE3')

# Plot bars for black people
plt.bar(positions + width/3, descent_counts_by_area['Vict Descent_B'], width=width, label='Black', color='#A9DFBF')

# Plot bars for hispanic people
plt.bar(positions + width/3, descent_counts_by_area['Vict Descent_H'], width=width, label='Hispanic', color='#D7BDE2')

plt.xlabel('AREA')
plt.ylabel('Number of Victims')
plt.title('Number of Victims by Descent and Area')
plt.xticks(positions, descent_counts_by_area['AREA'].astype(str), rotation=45)
plt.legend()

plt.show()
```
